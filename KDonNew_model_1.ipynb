```python
import torch
from torch import nn
import matplotlib.pyplot as plt
from transformers import BertForSequenceClassification, BertTokenizer

# Check if a GPU is available and set the device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load the teacher model (BERT Large) and move it to the device
teacher_model = BertForSequenceClassification.from_pretrained('bert-large-uncased')
teacher_model.to(device)

# Load the student model and move it to the device
student_model = BertForSequenceClassification.from_pretrained(output_dir)
student_model.to(device)

# Load the tokenizer
tokenizer = BertTokenizer.from_pretrained(output_dir)

# Set the temperature for distillation
temperature = 2.0

# Define the loss function for knowledge distillation
def knowledge_distillation_loss(student_logits, teacher_logits, temperature):
    student_probs = nn.functional.softmax(student_logits / temperature, dim=-1)
    teacher_probs = nn.functional.softmax(teacher_logits / temperature, dim=-1)
    loss = nn.functional.kl_div(student_probs.log(), teacher_probs, reduction='batchmean')
    return loss

# Set the models to evaluation mode
teacher_model.eval()
student_model.train()

# Move the models to the GPU device
teacher_model.to(device)
student_model.to(device)

# Initialize lists to track loss and accuracy
train_loss_list = []
val_loss_list = []
accuracy_list = []

# Define the number of training epochs
num_train_epochs = 3

# Perform knowledge distillation
for epoch in range(num_train_epochs):
    train_loss = 0.0
    val_loss = 0.0
    total_train_samples = 0
    total_val_samples = 0
    correct_predictions = 0

    # Training loop
    for batch in train_dataset:
        # Tokenize the input batch and move it to the GPU device
        inputs = tokenizer(batch['sentence'], padding=True, truncation=True, return_tensors='pt')
        inputs = {k: v.to(device) for k, v in inputs.items()}

        # Forward pass through the teacher model
        with torch.no_grad():
            teacher_outputs = teacher_model(**inputs)

        # Forward pass through the student model
        student_outputs = student_model(**inputs)

        # Compute the knowledge distillation loss
        loss = knowledge_distillation_loss(student_outputs.logits, teacher_outputs.logits, temperature)

        # Perform backward pass and update student model's parameters
        student_model.zero_grad()
        loss.backward()
        # Add optimizer step if required

        train_loss += loss.item() * inputs['input_ids'].size(0)
        total_train_samples += inputs['input_ids'].size(0)

    
```

```python
import torch
from torch import nn
import matplotlib.pyplot as plt
from transformers import BertForSequenceClassification, BertTokenizer

# Load the student model from the saved path
student_model_path = '/content/drive/MyDrive/Student_Models/student_model.pt'
student_model = torch.load(student_model_path, map_location=device)
student_tokenizer = BertTokenizer.from_pretrained('bert-large-uncased')

# Set the device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
student_model.to(device)
student_model.eval()

# Initialize lists to track loss and accuracy
loss_list = []
accuracy_list = []

# Iterate over your dataset
for batch in dataset:
    # Tokenize the input batch using the student tokenizer and move it to the device
    inputs = student_tokenizer(batch['sentence'], padding=True, truncation=True, return_tensors='pt')
    inputs = {k: v.to(device) for k, v in inputs.items()}

    # Forward pass through the student model
    with torch.no_grad():
        outputs = student_model(**inputs)

    # Calculate the loss
    loss = nn.CrossEntropyLoss()(outputs.logits, inputs['labels'])
    loss_list.append(loss.item())

    # Calculate accuracy
    predicted_labels = torch.argmax(outputs.logits, dim=1)
    correct_predictions = (predicted_labels == inputs['labels']).sum().item()
    accuracy = correct_predictions / inputs['labels'].size(0)
    accuracy_list.append(accuracy)

# Calculate average loss and accuracy
average_loss = sum(loss_list) / len(loss_list)
average_accuracy = sum(accuracy_list) / len(accuracy_list)

# Plot the loss and accuracy graphs
epochs = range(1, len(loss_list) + 1)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs, loss_list, 'b-', label='Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epochs, accuracy_list, 'g-', label='Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training Accuracy')
plt.legend()

plt.tight_layout()
plt.show()

print(f"Average Loss: {average_loss}")
print(f"Average Accuracy: {average_accuracy}")
```
